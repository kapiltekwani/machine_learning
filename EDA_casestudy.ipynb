```python
#Next delete mths_since_last_delinq, mths_since_last_record, next_pymnt_d
#As there is more than 50% columns are having null values

del loan['next_pymnt_d']
del loan['mths_since_last_delinq']
del loan['mths_since_last_record']
```

```python
#record the columns which have only 1 value in it
unique_list = []
for i in loan.columns:
    if loan[i].nunique() == 1:
        unique_list.append(i)
unique_list
```

```python
loanCleaned = loan.drop(columns=unique_list)
loanCleaned.shape
```

```python
loanCleaned.to_csv('loan_cleaned.csv')
```
